```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
poll_data = pd.read_csv('kicker.csv')
```

```python
plt.scatter(poll_data['red_team_goals'], poll_data['yellow_team_goals'])
plt.xlabel('Red team goal')
plt.ylabel('Yellow team goals')
plt.title('Goals')
```

```python
sns.heatmap(poll_data.corr())
```

```python
sns.boxplot(x='red_won', y='red_team_goals', data=poll_data)
```

```python
sns.countplot(hue='red_team_p1_id', x='red_won', data=poll_data)
```

```python
poll_data['red_team_goals'].value_counts()
```
